# My Favorite Composers

```python
classical = spotify[(spotify['artist_name'] == 'Johann Sebastian Bach') |
        (spotify['artist_name'] == 'Ludwig van Beethoven') |
         (spotify['artist_name'] == 'Wolfgang Amadeus Mozart') |
         (spotify['artist_name'] == 'Sergei Rachmaninoff')]
         # Here I filtered for four of my favorite composers

classical
```

```python
classical.set_index('artist_name', inplace = True)
 # Here I set the index to artist_name
  # so as to better locate each composer's information in the data set
```

## Johann Sebastian Bach

```python
jsb = (classical.loc[['Johann Sebastian Bach']])
# I used the .loc[] accessor to access only obs. for Johann Sebastian Bach,
  # and will do the same in the sections to follow for the respective composers
jsb.drop_duplicates(subset = ['track_name'], keep = 'first')
```

```python
jsb['track_name'].nunique()
# There are 50 unique tracks by J.S. Bach
```

```python
(
    jsb[['track_name','duration_ms']]
    .sort_values(['duration_ms'], ascending = False)
    .head()
) # Here are the 5 longest pieces by J.S. Bach
```

## Ludwig van Beethoven

```python
lvb = (classical.loc[['Ludwig van Beethoven']])
lvb.drop_duplicates(subset = ['track_name'], keep = 'first')
```

```python
lvb['track_name'].nunique()
 # There are 25 pieces by Beethoven in the data set
```

```python
(
    lvb[['track_name','duration_ms']]
    .sort_values(['track_name'], ascending = False)
    .head()
) # Here are the 5 longest pieces by Beethoven
```
